# Moonshot Backtest

We're ready to run a backtest. Moonshot returns the backtest results as a CSV, and we'll use Moonchart, a companion library to Moonshot, to look a tear sheet of performance results.

First, run the backtest:

```python
from quantrocket.moonshot import backtest
backtest("return-momentum", start_date="2019-01-01", end_date="2023-01-12", 
         params = {
             "MOMENTUM_WINDOW": 250, "RANKING_PERIOD_GAP": 10, "TOP_N_PCT": 8, "TOP_N_COUNT": None, 
             "REBALANCE_INTERVAL": "M", "PRICE_LOWER_LIMIT": 800,
             "LONG_ONLY": True, "ESCAPE_WEEKLY_CHANGE_LIMIT": 0.1,
             "DB": "usstock-1d", "UNIVERSES": "usstock"},
         filepath_or_buffer="umd_moonshot_results.csv")
```

Then, view a tear sheet:

```python
from moonchart import Tearsheet
Tearsheet.from_moonshot_csv("return_momentum_moonshot_results.csv")
```

```python
from quantrocket.moonshot import scan_parameters
scan_parameters("return-momentum", 
                param1="LONG_ONLY",
                vals1=[True, False],
                param2="ESCAPE_WEEKLY_CHANGE_LIMIT",
                vals2=[0.0, 0.1, 0.2],
                params = {
                   "MOMENTUM_WINDOW": 250, "RANKING_PERIOD_GAP": 10, "TOP_N_PCT": None, "TOP_N_COUNT": 4, 
                   "REBALANCE_INTERVAL": "M", "PRICE_LOWER_LIMIT": 500, 
                    "LONG_ONLY": True, "ESCAPE_WEEKLY_CHANGE_LIMIT": 0.2,
                   "DB": "usstock-1d", "UNIVERSES": "usstock"},
                start_date="2019-01-01", 
                end_date="2023-01-12", 
                filepath_or_buffer="umd_moonshot_ESCAPE_LONG.csv")
```

```python
from moonchart import ParamscanTearsheet
ParamscanTearsheet.from_csv("return_momentum_moonshot_ESCAPE_LONG.csv")
```
